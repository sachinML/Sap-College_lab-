```python
import scipy.io.wavfile  as sw
import IPython.display as ipd
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
ipd.Audio('recorded.wav')
```

```python
fs,y = sw.read('recorded.wav')
print(fs)
print(y)
plt.plot(y)
#ipd.display(ipd.Audio(data = y,rate = fs))
```

```python
b, a = signal.butter(8, 3400,'low',fs = 16000)
b,a

w, h = signal.freqz(b, a)
#w*=fs/(2*np.pi)
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid(which='both', axis='both')
plt.show()
```

```python
y1 = signal.lfilter(b, a, y)
ipd.display(ipd.Audio(data=y1, rate=fs))
```

```python
b1, a1 = signal.butter(8, 300, 'hp',fs = 16000)
b1,a1

w1, h1 = signal.freqz(b1, a1)
#w1*= fs/(2*np.pi)
plt.semilogx(w1, 20 * np.log10(abs(h1)))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid(which='both', axis='both')
plt.show()
```

```python
y2 = signal.lfilter(b1, a1, y1)
```

```python
ipd.display(ipd.Audio(data=y2, rate=fs))
```
