```python
import matplotlib.pyplot as plt
import numpy as np

x = plt.imread('airplane.png')

plt.imshow(x)
plt.show()
x.shape
```

```python
import scipy.signal as sig

def convimage(x,h):
    y = np.zeros(x.shape)
    for i in range(x.shape[2]):
        y[:,:,i] = sig.convolve2d(x[:,:,i],h,'same') #'same'
    mx = np.max(y)
    mn = np.min(y)
    y = (y - mn)/(mx - mn)
    return y
    
h1 = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
h2 = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
h3 = np.array([[1,2,1],[2,4,2],[1,2,1]])

y11 = convimage(x,h1)
y12 = convimage(x,h2)
y13 = convimage(x,h3)



fig = plt.figure(figsize = (12,10))
fig.add_subplot(1,3,1)
plt.imshow(y11)
fig.add_subplot(1,3,2)
plt.imshow(y12)
fig.add_subplot(1,3,3)
plt.imshow(y13)
```

```python
def convimage3(x,g,f):
    y = np.zeros(x.shape)
    for i in range(x.shape[2]):
        y[:,:,i] = np.square(sig.convolve2d(x[:,:,i],g,'same')) + np.square(sig.convolve2d(x[:,:,i],f,'same'))
    mx = np.max(y)
    mn = np.min(y)
    y = (y - mn)/(mx - mn)
    return y
    
y41 = convimage3(x,h1,h2)
plt.imshow(y41)
```

```python
def convimage31(x,g,f):
    y = np.zeros(x.shape)
    for i in range(x.shape[2]):
        y[:,:,i] = np.sqrt(np.square(sig.convolve2d(x[:,:,i],g,'same')) + np.square(sig.convolve2d(x[:,:,i],f,'same')))
    mx = np.max(y)
    mn = np.min(y)
    y = (y - mn)/(mx - mn)
    return y
    
y412 = convimage31(x,h1,h2)
plt.imshow(y412)
```

```python
def convimage4(x,g):
    y = np.zeros(x.shape)
    for i in range(x.shape[2]):
        y[:,:,i] = x[:,:,i] - sig.convolve2d(x[:,:,i],g,'same') 
    mx = np.max(y)
    mn = np.min(y)
    y = (y - mn)/(mx - mn)
    return y
    
y5 = convimage4(x,h3)
plt.imshow(y5)
```
