```python
import numpy as np 
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt

# Set the sampling rate
rate = 8000;

# Genearate the t values use np.arange
t = np.arange(0,0.2,1.0/8000)

#Generate signal 
s1 = np.sin(100*np.pi*t)
s2 = np.cos(1000*np.pi*t)
s3 = np.sin(100*np.pi*np.square(t))

#Write thee signal 
wav.write('sin100.wav',rate, s1);
plt.stem(t[0:100],s1[0:100])
plt.show()

wav.write('cos100.wav',rate, s2);
plt.stem(t[0:100],s2[0:100])
plt.show()

wav.write('chirp.wav',rate, s3);
plt.stem(t[0:100],s3[0:100])
plt.show()
```
