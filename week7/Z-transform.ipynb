Q1. Plot the pole zero locations and time domain impulse response (clearly mark the point n=0 ) for the following transfer functions. Assume all signals are causal. If the DTFT exists, plot the magnitude, phase and group delay from $-\pi$ to $\pi$.
<ol>
  <li>$ \frac{1}{1+az^{-1}} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
 <li>$ \frac{1}{(1+az^{-1})^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
  <li>$\frac{1}{1-az^{-1}+{(az^{-1})}^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Complex conjugate poles) </li>    
</ol>
Which of the responses corresponds to stable LTI systems?

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.tf2zpk.html

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

def plot_pole_zero(b,a):
    [z,p,k] = sig.tf2zpk(b,a)   #zero,pole,system gain  #numerator, denominator representation of a linear filter
    t1 = plt.plot(p.real, p.imag, 'rx')
    t2 = plt.plot(z.real, z.imag, 'b.')
    plt.setp(t1, markersize=12.0, markeredgewidth=3.0)
    plt.setp(t2, markersize=12.0, markeredgewidth=3.0)
    plt.show()
    
def plot_dimpulse(b,a,t):
    t,y = sig.dimpulse((b,a,1),n = t)
    plt.stem(t,y[0])
    plt.ylabel('Impulse Response')
    plt.show()
    
def plot_group_delay(b,a):
    w, gd = sig.group_delay((b, a))
    plt.title('Digital filter group delay')
    plt.plot(w, gd)
    plt.ylabel('Group delay [samples]')
    plt.xlabel('Frequency [rad/sample]')
    plt.show()

def plot_response(b, a, c):
    w1,h1=sig.freqz(b,a,c)
    h1db=20*np.log10(abs(h1))
    angles = np.angle(h1)
    fig = plt.figure(figsize = (15,5))
    fig.add_subplot(1,2,1)
    plt.title('Magnitude Response')
    plt.plot(w1/(2*np.pi),h1db)
    fig.add_subplot(1,2,2)
    plt.title('Phase Response')
    plt.plot(w1/(2*np.pi), angles, 'g')
    plt.show()
```

```python
#1
#a = 0.8
b1=[1,0]
a1=[1,0.8]
c=np.arange(-2*np.pi, 2*np.pi, 4*np.pi/4096)
plot_pole_zero(b1,a1)
plot_dimpulse(b1,a1,10)
plot_response(b1,a1,c)
plot_group_delay(b1,a1)

#a = -0.8
b1=[1,0]
a1=[1,-0.8]
c=np.arange(-2*np.pi, 2*np.pi, 4*np.pi/4096)
plot_pole_zero(b1,a1)
plot_dimpulse(b1,a1,10)
plot_response(b1,a1,c)
plot_group_delay(b1,a1)
```
